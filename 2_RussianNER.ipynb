```python
text = 'КПРФ решила передать место Алферова в Госдуме Грудинину'
```

```python
from pullenti_wrapper.processor import (
    Processor,
    DATE,
    GEO,
    ORGANIZATION,
    PERSON,
    MONEY,
)

processor = Processor([PERSON, ORGANIZATION, GEO, DATE, MONEY])

result = processor(text)
display(result)
```

```python
result.graph
```

```python
print(result.matches[1].referent.label)
print(text[result.matches[1].span.start:result.matches[1].span.stop])
print(result.matches[1].referent.slots[2].value)
```

```python
from natasha import*

all_extractors = [NamesExtractor, SimpleNamesExtractor, DatesExtractor, MoneyExtractor,
              MoneyRateExtractor, MoneyRangeExtractor, LocationExtractor, AddressExtractor,
              OrganisationExtractor, PersonExtractor]


extractors = [extractor() for extractor in all_extractors]

for extractor in extractors:
    matches = extractor(text)
    for match in matches:
        print(match.span, match.fact)
```
